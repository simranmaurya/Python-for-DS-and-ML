```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Functional
x = np.arange(0,10)
y = x**2
plt.title('Num vs Sqr')
plt.xlabel('Numbers')
plt.ylabel('Number Sqaure')
plt.plot(x,y)
plt.show()
```

```python
plt.subplot(2,2,1)
plt.plot(x,y,'r')

plt.subplot(2,2,2)
plt.plot(x,y,'b')

plt.subplot(2,2,3)
plt.plot(x,y,'g')

plt.subplot(2,2,4)
plt.plot(x,y,'y')
```

```python
# Object oriented
fig = plt.figure(figsize=(2,2))

ax1 = fig.add_axes([0.1,0.1,1,1])
# ax2 = fig.add_axes([0.1,0.1,0.5,0.5])

ax1.plot(x,y,'r')
ax1.plot(y,x,'b')
plt.show()
```

```python
fig.savefig('graph.jpeg',dpi=1000)
fig.savefig('graph2.jpg',dpi=2000)
```

```python
fig,axes = plt.subplots(1,2)

for current in axes:
    current.plot(x,y)
plt.tight_layout
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = 'X2',lw = 3,ls = '-', marker = 'o',markersize = 10,markeredgewidth = 2, markeredgecolor = 'black')
axes.plot(x,x**3,label = 'X3', lw = 3,ls = '-', marker = '*',markersize = 15,markeredgewidth = 2,markeredgecolor = 'black')
axes.legend()
```

```python
fig,axes = plt.subplots(1,2,figsize = (12,5))

x = np.arange(0,5)
axes[0].plot(x,x+1,'orange',x,x+2,'blue',x,x+3,'green')
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=2)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18); # use LaTeX formatted labels
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

from matplotlib import ticker
formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True) 
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter) 
```
