## Matrix Plots

```python
tips = sns.load_dataset('tips')
flights = sns.load_dataset('flights')
```

```python
tips.head()
```

```python
flights.head()
```

```python
tc = tips.corr()
fc = flights.corr()
```

```python
tc
fc
```

```python
sns.heatmap(tc, annot = True)
```

```python
sns.heatmap(fc, annot = True)
```

```python
fp = flights.pivot_table(index = 'month', columns = 'year', values = 'passengers')
```

```python
sns.heatmap(fp, linecolor = 'white', linewidth = 1)
```

```python
tips.head()
```

```python
tp = tips.pivot_table(index = 'day',columns = 'sex', values = ['total_bill', 'tip'])
```

## Grids

```python
sns.pairplot(tips)
```

```python
g = sns.PairGrid(tips)
g.map_diag(sns.boxplot)
g.map_upper(sns.kdeplot)
g.map_lower(plt.scatter)
```

```python
m = sns.FacetGrid(data = tips, col = 'sex', row = 'day')
m.map(sns.kdeplot,'tip')
```
